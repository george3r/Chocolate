```python
from bs4 import BeautifulSoup 
import requests
```

```python
chocolate_alchemy_url = 'https://chocolatealchemy2.myshopify.com/collections/cocoa-beans'
```

```python
r = requests.get(chocolate_alchemy_url)
```

```python
html_doc = r.text
```

```python
soup = BeautifulSoup(html_doc)
```

```python
pretty_soup = soup.prettify()
```

```python
title = soup.title
text = soup.text
```

```python
text_organized = text.split('\n')
```

```python
text_organized = [i for i in text_organized if i]
```

```python
#text_organized
```

```python
### MUST INPUT THE FIRST AND LAST BEAN NAMES ON THE WEBSITE ###
first_bean = 'Belize Organic - 2020'
last_bean = 'Uganda Semuliki Forest 2020 Direct Trade'
```

```python
include = False
beans = []
for i in text_organized:
    if i == first_bean:
        include = True

    if include == True:
        beans.append(i)
        
    if i == last_bean:
        include = False
```

```python
beans
```

```python
for c,bean in enumerate(beans):
    beans[c] = bean.replace('/','-').replace('ñ','n').replace('ó','o').lower()
```

```python
beans_hyperlinks = []
for i in beans:
    beans_hyperlinks.append(chocolate_alchemy_url+'/products/'+i)
for c,i in enumerate(beans_hyperlinks):
        beans_hyperlinks[c] = i.replace(' ','-')
for c,i in enumerate(beans_hyperlinks):
        beans_hyperlinks[c] = i.replace('---','-')
for c,i in enumerate(beans_hyperlinks):
        beans_hyperlinks[c] = i.replace('---','-')
for c,i in enumerate(beans_hyperlinks):
        beans_hyperlinks[c] = i.replace('--','-')
```

```python
beans_hyperlinks
```

```python
url = beans_hyperlinks[0]
page = requests.get(url)
html_doc_page = page.text
url_soup = BeautifulSoup(html_doc_page)
```
